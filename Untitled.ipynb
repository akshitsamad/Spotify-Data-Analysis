1. IMPORTING PYTHON LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. IMPORING DATA

```python
artists = pd.read_csv("artists.csv")
tracks = pd.read_csv("tracks.csv")
```

```python
tracks.head()
```

```python
artists.head()
```

3. IDENTIFYING NULL VALUES IN DATASET

```python
pd.isnull(artists).sum()
```

```python
pd.isnull(tracks).sum()
```

4. REMOVING THOSE NULL VALUES

```python
tracks = tracks.dropna(subset=['name'])

```

```python
artists = artists.dropna(subset=['name', 'followers'])
```

```python
pd.isnull(tracks).sum()
```

```python
pd.isnull(artists).sum()
```

5. DATASET OVERVIEW

```python
artists.info()
```

```python
tracks.info()
```

6. EXPLORING THE 10 LEAST POPULAR SONGS IN DATASET


```python
least_popular = tracks.sort_values(by='popularity', ascending= True).head(10)[['name', 'popularity', 'artists']].reset_index(drop=True)
least_popular
```

7. DESCRIPTIVE STATISTICS

```python
tracks.describe().transpose()
```

8. DISCOVERING THE TOP 10 POPULAR SONGS IN THE DATASET

```python
top_10 = tracks.sort_values(by='popularity', ascending = False).head(10)[['name', 'popularity', 'artists']].reset_index(drop=True)
top_10
```

9. SETTING RELEASE DATE AS THE INDEX COLUMN

```python
tracks.set_index('release_date', inplace=True)
```

10. CONVERTING SONG DURATION FROM MILLISECONDS TO SECONDS 

```python
tracks['duration_sec'] = (tracks['duration_ms']/1000)
tracks['duration_sec'].head()
```

11. VISUALIZATION: PEARSON CORRELATIONHEATMAP FOR TWO VARIABLES

```python
corr = tracks[["popularity", "duration_sec"]].corr(method="pearson")


plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()
```

12. CREATING 4% SAMPLE OF THE ENTIRE DATASET

```python
sample_4_percent = tracks.sample(frac=0.04, random_state=42)
sample_4_percent
```

13. REGRESSION PLOT OF LOUDNESS VS. ENERGY WITH REGRESSION LINE 

```python
plt.figure(figsize=(8,6))
sns.regplot(x='loudness', y='energy', data=sample_4_percent, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Regression Plot of Loudness vs Energy')
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```

14. REGRESSION PLOT OF POPULARITY VS. ACOUSTICNESS WITH REGRESSION LINE 

```python
plt.figure(figsize=(8,6))
sns.regplot(x='popularity', y='acousticness', data= sample_4_percent, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Regression Plot of Popularity vs Acousticness')
plt.xlabel('Popularity')
plt.ylabel('Acousticness')
plt.show()
```

15. ADDING YEAR COLUMN


```python
tracks.index = pd.to_datetime(tracks.index, errors='coerce')
tracks.index_name = 'release_date'
tracks['year'] = tracks.index.year
tracks.head()
```

16. GRAPH: NUMBER OF SONGS PER YEAR

```python
songs_per_year = tracks['year'].value_counts().reset_index()
songs_per_year
```

```python
sns.displot(data=songs_per_year, x='year', weights= 'count', bins=len(songs_per_year))
plt.title('Number of Songs per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```
